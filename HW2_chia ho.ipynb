### f. use daily returns of Bitcoin and the Altcoin you choose and run the following regression: 𝐴𝑙𝑡𝑐𝑜𝑖𝑛_𝑅𝑒𝑡+ = 𝐶𝑜𝑛𝑠𝑡𝑎𝑛𝑡 + 𝛽 ́ 𝐵𝑇𝐶_𝑅𝑒𝑡+ + 𝑒
### where 𝐴𝑙𝑡𝑐𝑜𝑖𝑛_𝑅𝑒𝑡+ is the Altcoin return on date 𝑡, and 𝐵𝑇𝐶_𝑅𝑒𝑡+ is the Bitcoin return on date 𝑡.
### Report the regression result, including 𝐶𝑜𝑛𝑠𝑡𝑎𝑛𝑡, 𝛽, t-stats, and R-squared. Comment on the estimated coefficients in terms of statistical significance and economic magnitude. Can Altcoin returns be explained by Bitcoin returns? [5 points]

```python
df_bitcoin_match.reset_index(drop=True, inplace=True)
bitcoin_ret = df_bitcoin_match['daily return (%)']
```

```python
df_sol.reset_index(drop=True, inplace=True)
sol_ret = df_sol['daily return (%)']
```

```python
bitcoin_ret = bitcoin_ret.dropna()
```

```python
sol_ret = sol_ret.dropna()
```

```python
import statsmodels.api as sm

# Assuming you have two variables bitcoin_ret and sol_ret

# Create the regression model
model = sm.OLS(sol_ret, sm.add_constant(bitcoin_ret.values.reshape(-1, 1)))

# Fit the model
result = model.fit()

# Get the coefficients and intercept
beta = result.params[1]
constant = result.params[0]
r_squared = result.rsquared

# Print the regression results
print("Intercept (Constant):", constant)
print("Coefficient (Beta):", beta)
print("R-squared:", r_squared)
```

```python
print(result.summary())
```

In the regression model, the r square is relatively low with only .2, which shows that the returns of the Solana can not be explained by bitcoin's return.
