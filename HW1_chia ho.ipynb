#### Q6. use daily returns of Bitcoin and run the following regressions:
Report the results of regression Equation (1) in Table 3, including , 0, t-stats, and R- squared (see the template table below). In addition, run five similar regressions with 2, 3, 4, 5, and 6 lags using Equations (2) to (6). Add the results of these regression to Table 3 (i.e., Table 3 has results of 6 regression models). Comment on the estimated coefficients in terms of statistical significance and economic magnitude. Can Bitcoin return be explained by its past returns? (You do not need to explain why, just state the fact based on the regression results) [5 points]

```python
df = pd.DataFrame(bitcoin_returns)
df
```

```python

# Create lags
for lag in range(1,7):
    df[f'BTCRe_t({lag})'] = df['daily return (%)'].shift(lag)

print(df)
```

```python
# Create lag variables
df['BTCRet(t-1)'] = df['daily return (%)'].shift(1)
df['BTCRet(t-2)'] = df['daily return (%)'].shift(2)
df['BTCRet(t-3)'] = df['daily return (%)'].shift(3)
df['BTCRet(t-4)'] = df['daily return (%)'].shift(4)
df['BTCRet(t-5)'] = df['daily return (%)'].shift(5)
df['BTCRet(t-6)'] = df['daily return (%)'].shift(6)


# Handle missing values (NaN) due to lag creation
df.dropna(inplace=True)  # Choose an appropriate way to handle missing values

```

```python
import statsmodels.api as sm

models = []  # Store regression results

for lags in range(1,7,1): # Run regressions for 1 to 6 lags
    X = df[f'BTCRet(t-{lags})']  # Select the lag as the predictor
    X = sm.add_constant(X) 
    y = df['daily return (%)']

    model = sm.OLS(y, X).fit()
    models.append(model)
```

```python
import pandas as pd

# set up an empty DataFrame
results_table = pd.DataFrame(columns=['1', '1 t-stat', '2', '2 t-stat',
                                      '3', '3 t-stat', '4', '4 t-stat',
                                      '5', '5 t-stat', '6', '6 t-stat',
                                      'Constant', 'Constant t-stat', 'R-squared'])

for i, model in enumerate(models):
    row = {}  # use dictionary to store data
    row['1'] = model.params.get(f'BTCRet(t-{i + 1})', '')
    row['1 t-stat'] = model.tvalues.get(f'BTCRet(t-{i + 1})', '')
    # add  and tstats value
    row['Constant'] = model.params.get('const', '')
    row['Constant t-stat'] = model.tvalues.get('const', '')
    row['R-squared'] = model.rsquared

    # add rows into results_table
    results_table = pd.concat([results_table, pd.DataFrame(row, index=[0])], ignore_index=True)

print(results_table)
```

```python
for i, model in enumerate(models):
    print(f"Model {i+1} Results:")
    print("1:", model.params.get(f'BTCRet(t-{i + 1})', ''))
    print("1 t-stat:", model.tvalues.get(f'BTCRet(t-{i + 1})', ''))
    
    # Add more coefficients and statistics as needed
    print("Constant:", model.params.get('const', ''))
    print("Constant t-stat:", model.tvalues.get('const', ''))
    print("R-squared:", model.rsquared)
    print("------------------------------------")
```

According to the results, the R-square for all the models are low. Therefore, the bitcoin price cannot be explained in it's past pattern.
